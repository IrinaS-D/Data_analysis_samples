```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import copy

df = pd.read_csv("Emdat1_public_2023_02_14_query_uid-QKBw5b.csv")
```

```python
df
```

```python
df1 = df[["Year", "Country", "Location", "River Basin","Disaster Group", "Disaster Subgroup", "Disaster Type", "Disaster Subtype", "Event Name", "Origin", "Associated Dis", "Start Month", "No Affected","No Homeless", "Total Deaths", "Total Affected", "Reconstruction Costs, Adjusted ('000 US$)", "Insured Damages ('000 US$)", "Insured Damages, Adjusted ('000 US$)", "Total Damages ('000 US$)", "Total Damages, Adjusted ('000 US$)"]]
df1
```

```python
df2 = df1[["Country", "Total Affected"]]
df2
```

```python
df_affected = df2.groupby("Country").sum()
print(df_affected)
```

```python
df_total = df_affected.sort_values(by="Total Affected", ascending=False)
df_total
```

```python
df_10 = df_total.head(10)
df_10
```

```python
df_join = df_10.join(df_cpop.set_index('Country'), on='Country')
print(df_join)
```

```python
df_join.plot(kind = "bar")
```
