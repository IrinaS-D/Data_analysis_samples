```python
df_1965 = df2.head(5).groupby("Date").max()
```

```python
plt.plot(df_1965)
plt.title("Magnitude by Date - 1965", fontsize = 14, fontweight = "bold", color = "green")
plt.xlabel("Date", fontsize = 11, color = "#006666")
plt.ylabel("Magnitude", fontsize = 11, color = "#006666")
plt.show()
```

```python
height = 1
plt.bar(df_1965, height=1)
plt.title("Magnitude by Date - 1965", fontsize = 14, fontweight = "bold", color = "green")
plt.xlabel("Date", fontsize = 11, color = "#006666")
plt.ylabel("Magnitude", fontsize = 11, color = "#006666")
plt.show()
```
